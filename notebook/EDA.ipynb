```python
# Eliminamos StudentID porque no aporta información
# Eliminamos GradeClass porque es una variable categórica que no aporta información para la predicción del GPA

# Variable objetivo
y = df["GPA"]

# Variables predictoras (sin StudentID, GradeClass ni GPA)
X = df.drop(columns=["StudentID", "GradeClass", "GPA"])
```

Se hará una división del 80 % para entrenamiento y 20 % para pruebas

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print(X_train.shape)
print(X_test.shape)
```

Primer modelo (Regresión Lineal)

```python

from sklearn.linear_model import LinearRegression

modelo_lr = LinearRegression()

modelo_lr.fit(X_train, y_train)

pred_lr = modelo_lr.predict(X_test)
```

```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
import numpy as np

mae = mean_absolute_error(y_test, pred_lr)
rmse = np.sqrt(mean_squared_error(y_test, pred_lr))
r2 = r2_score(y_test, pred_lr)

print("Regresión Lineal")
print("MAE:", round(mae,4))
print("RMSE:", round(rmse,4))
print("R²:", round(r2,4))
```

Segundo Modelo: Árbol de Decisión

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
modelo_dt = DecisionTreeRegressor(
    random_state=42
)

modelo_dt.fit(X_train, y_train)

pred_dt = modelo_dt.predict(X_test)
```

```python
mae_dt = mean_absolute_error(y_test, pred_dt)
rmse_dt = np.sqrt(mean_squared_error(y_test, pred_dt))
r2_dt = r2_score(y_test, pred_dt)

print("Árbol de Decisión")
print("MAE:", round(mae_dt,4))
print("RMSE:", round(rmse_dt,4))
print("R²:", round(r2_dt,4))
```

Tercer Modelo: Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
modelo_rf = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

modelo_rf.fit(X_train, y_train)

pred_rf = modelo_rf.predict(X_test)
```

```python
mae_rf = mean_absolute_error(y_test, pred_rf)
rmse_rf = np.sqrt(mean_squared_error(y_test, pred_rf))
r2_rf = r2_score(y_test, pred_rf)

print("Random Forest")
print("MAE:", round(mae_rf,4))
print("RMSE:", round(rmse_rf,4))
print("R²:", round(r2_rf,4))
```

Comparar modelos

```python
resultados = pd.DataFrame({
    "Modelo": [
        "Regresión Lineal",
        "Árbol de Decisión",
        "Random Forest"
    ],
    "MAE": [
        mae,
        mae_dt,
        mae_rf
    ],
    "RMSE": [
        rmse,
        rmse_dt,
        rmse_rf
    ],
    "R²": [
        r2,
        r2_dt,
        r2_rf
    ]
})

resultados
```

```python
importancias = pd.DataFrame({
    "Variable": X.columns,
    "Importancia": modelo_rf.feature_importances_
})

importancias = importancias.sort_values(
    by="Importancia",
    ascending=False
)

importancias
```

Según estos resultados, se concluye que el modelo con Regresión Lineal es más adecuado para este analisis.

```python
#Validación cruzada para el modelo de regresión lineal
from sklearn.model_selection import cross_val_score

scores = cross_val_score(
    LinearRegression(),
    X,
    y,
    cv=5,
    scoring="r2"
)

print(scores)
print("R² promedio:", scores.mean())
```

Para evaluar la capacidad de generalización del modelo, se aplicó una validación cruzada de cinco particiones (5-Fold Cross Validation). El modelo obtuvo un coeficiente de determinación promedio (R²) de 0.9534, mostrando resultados consistentes entre las diferentes particiones. Esto indica que la Regresión Lineal presenta un buen poder predictivo y una adecuada capacidad de generalización sobre datos no vistos.

```python
plt.figure(figsize=(7,7))

plt.scatter(y_test, pred_lr, alpha=0.6)

plt.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    'r--'
)

plt.xlabel("GPA Real")
plt.ylabel("GPA Predicho")
plt.title("Predicción vs Valor Real")

plt.show()
```

```python
errores = y_test - pred_lr

plt.figure(figsize=(8,5))

sns.histplot(errores, kde=True)

plt.title("Distribución de los errores")

plt.show()
```

La distribución de los errores muestra una concentración alrededor de cero y una forma aproximadamente normal, lo que indica que el modelo no presenta un sesgo sistemático hacia la sobreestimación o subestimación del GPA. Asimismo, la mayoría de los errores son de baja magnitud, observándose únicamente algunos valores atípicos aislados.

```python
residuos = y_test - pred_lr

plt.figure(figsize=(8,5))

plt.scatter(pred_lr, residuos, alpha=0.6)

plt.axhline(0, color="red", linestyle="--")

plt.xlabel("GPA Predicho")
plt.ylabel("Residuales")
plt.title("Residuales vs Valores Predichos")

plt.show()
```
